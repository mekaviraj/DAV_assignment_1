```python
import pandas as pd

# 🎬 Movie stats dataset
df_stats = pd.DataFrame({
    "IMDB_Rating": [8.5, 9.0, 8.7],
    "Runtime": [142, 152, 130],
    "Votes": [2000000, 1500000, 1800000]
}, index=["Inception", "The Dark Knight", "Interstellar"])

# 🎭 Movie genre dataset
df_genre = pd.DataFrame({
    "Genre_1": ["Sci-Fi", "Action", "Sci-Fi"],
    "Genre_2": ["Thriller", None, "Drama"]
}, index=["Inception", "The Dark Knight", "Interstellar"])

# 1. Default Join
df_joined = df_stats.join(df_genre)
print("\n🔹 Default Join:\n", df_joined)

# 2. INNER JOIN
df_inner = df_stats.join(df_genre, how="inner")
print("\n🔹 INNER JOIN:\n", df_inner)

# 3. OUTER JOIN
df_outer = df_stats.join(df_genre, how="outer")
print("\n🔹 OUTER JOIN:\n", df_outer)

# 4. RIGHT JOIN
df_right = df_stats.join(df_genre, how="right")
print("\n🔹 RIGHT JOIN:\n", df_right)
```
